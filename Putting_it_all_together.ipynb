```python
import pandas as pd
```

```python
country_df = pd.read_csv('GlobalLandTemperaturesByCountry.csv')
country_df
```

```python
long_lat_df = pd.read_csv('countries.csv')
long_lat_df
```

```python
long_lat_df= long_lat_df.drop(['Importance','Altitude'], axis=1)
```

```python
long_lat_df = long_lat_df[long_lat_df['Latitude']>= 0]
long_lat_df
```

```python
countries_df = country_df.merge(long_lat_df)
```

```python
countries_df = countries_df.dropna()
```

```python
countries_df['dt'] = pd.to_datetime(countries_df['dt'], format='%Y-%m-%d')

filtered_countries_df = countries_df.loc[(countries_df['dt'] >= '1900-01-01')]

filtered_countries_df
```
