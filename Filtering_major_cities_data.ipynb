```python
import pandas as pd
```

```python
#load in dataframe
GLTbymcity_df = pd.read_csv('GlobalLandTemperaturesByMajorcity.csv')
GLTbymcity_df
```

```python
#observing the data
GLTbymcity_df.dtypes
```

```python
GLTbymcity_df['Latitude'] = GLTbymcity_df['Latitude'].astype('str')
```

```python
#observing the data
GLTbymcity_df.dtypes
```

```python
#observing the data
GLTbymcity_df.City.unique()
```

```python
#observing the data
GLTbymcity_df.Country.unique()
```

```python
#observing the data
GLTbymcity_df.Country.value_counts()
```

```python
#filtering data for only northern hemisphere
GLTbymcity_df = GLTbymcity_df[GLTbymcity_df['Latitude'].str.contains('N')]
GLTbymcity_df
```

```python
#looking for empty rows
GLTbymcity_df.isnull().sum()
```

```python
#removing empty rows
GLTbymcity_df.dropna()
```

```python
#using loc to filter data
#url used https://www.geeksforgeeks.org/how-to-filter-dataframe-rows-based-on-the-date-in-pandas/#:~:text=To%20filter%20rows%20based%20on,to%20specify%20a%20filter%20condition.

GLTbymcity_df['dt'] = pd.to_datetime(GLTbymcity_df['dt'], format='%Y-%m-%d')

filter_df = GLTbymcity_df.loc[(GLTbymcity_df['dt'] >= '1900-01-01')]

filter_df
```

```python
filter_df.dropna()
```
